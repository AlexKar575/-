Проанализировав график, можно говорить о том, что имеет место быть **правосторонняя ассиметрия**.

```python
sns.boxplot(y = 'Денежные доходы' , data = df)
```

```python
Q1 = int(df['Денежные доходы'].quantile(0.25))
Q3 = int(df['Денежные доходы'].quantile(0.75))
IQR=Q3-Q1
up_D= Q3 + 1.5*IQR
print(up_D)
down_D= Q1-1.5*IQR
print(down_D)
```

```python
print(len(df[(df['Денежные доходы']> up_D)]))

d=len(df[(df['Денежные доходы']> up_D)])/len(df) * 100
round(d,2)
```

На основе построенного ящика с усами можно сказать, что всего 7 значений по признаку * "денежные доходы" являются **аномальными**. Доля аномальных наблюдений из общего числа наблюдения составил *8.54%*

```python
sns.boxplot(y = 'Потребительские расходы' , data = df)

Q2 = int(df['Потребительские расходы'].quantile(0.25))
Q4 = int(df['Потребительские расходы'].quantile(0.75))
IQR1=Q4-Q2
up_D1= Q4 + 1.5*IQR1
print(up_D1)
down_D1= Q2-1.5*IQR1
print(down_D1)

print(len(df[(df['Потребительские расходы']> up_D1)]))
d1=len(df[(df['Потребительские расходы']> up_D1)])/len(df) * 100
round(d1,2)
```

На основе построенного ящика с усами можно сказать, что всего 4 значения по признаку * "потребительские расходы" являются **аномальными**. Доля аномальных наблюдений из общего числа наблюдения составил *4.88%*
